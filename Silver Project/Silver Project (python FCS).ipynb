### Rules to win ROCK PAPER Scissors -


 - R + P = P

 - P + R = P
 - P + S = S
 - S + R = R
 - S + P = S
 - R + R = T
 - P + P = T
 - S + S = T

```python
#Batch 5 _________silver project(Rock-Paper-Scissor)___________________

import random

tieCount = 0
playerCount = 0 
computerCount = 0 

print("!----------------------->>>> Starting The Rock-Paper-Scissor Game <<<<-----------------------!")
player_name = input('\n\nPlease Enter Your Name: ')
print("-------------------------------------------------------------")

while True: 
    
    player_Choice = int(input("\n-> Pick a hand: \n   0: Rock\n   1: Paper\n   2: Scissor\n   3: Exit\n-> Please Enter a Number (0-3): "))
    
    if (player_Choice == 3):
        break 
    elif (player_Choice == 0 or player_Choice == 1 or player_Choice == 2):
        print("-------------------------------------------------------------")
        options = ['Rock','Paper','Scissor']
        computer_Choice = random.randint(0,2)
        print("  => "+player_name + " picked : " + options[player_Choice])
        print("  => Computer picked : " + options[computer_Choice])
        
        if player_Choice == computer_Choice:
            tieCount+=1
            print("\n -----> Result : Match Draw...")                   
        elif (player_Choice == 0 and computer_Choice == 1) or (player_Choice == 1 and computer_Choice == 2) or (player_Choice == 2 and computer_Choice == 0):
            computerCount+=2
            print("\n -----> Result : "+player_name+" Lose")           
        else:
            playerCount+=2
            print("\n -----> Result : "+player_name+" Winnnn...!!")
                                     
        print("\n!<====================== Score Board =====================>!")
        print("\n // *Get 2 Points Per Winning Match //")
        print("\n  => "+player_name+"'s Points : ",playerCount)
        print("  => Computer's Points : ",computerCount)
        print("  => Tie Match : ",tieCount)
        print('-------------------------------------------------------------')
        
    else:
        print("\n**Please Enter a Valid Number....")
```
